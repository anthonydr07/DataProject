```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot
from pylab import *
import numpy.linalg as linalg
from numpy import array
from scipy.linalg import svd
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns 
```

```python
dataset = pd.read_csv("diabetes.csv")
```

```python
# Récupération des données
X = dataset.iloc[:,0:8]
cols = ['Age','DiabetesPedigreeFunction','BMI','Insulin','SkinThickness','BloodPressure','Glucose','Pregnancies']
```

```python
W=np.corrcoef(X,rowvar=False)

s = sns.heatmap(X.corr(),cmap = 'coolwarm', annot = True)
s.set_yticklabels(s.get_yticklabels(),rotation = 30, fontsize = 10)
s.set_xticklabels(s.get_xticklabels(), rotation = 30, fontsize = 10)
plt.show()
```

```python
# définition de la catégorie outcome   
y = dataset.iloc[:,8]

# Standardisation pour avoir la moyenne = 0 et variance = 1
#(réalisation - sa moyenne)/(racine de la variance)
X_std = (X-np.mean(X,axis = 0))/np.std(X,axis = 0)
```

```python
# Etape 1: Trouver la matrice de covariance de X

# covariance de la matrice X 
cov_matrix = np.cov(X_std, rowvar=False) #fait X^{T}X/(n-1)
print(cov_matrix)
```

```python
# la matrice de covariance est une matrice nxn avec n=nombre de variable

# 2 étape 2: obtenir les valeurs propres et vecteurs propres
# valeurs propres et vecteurs propres
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# vecteurs propres
print('\nVecteurs propres de Cov(X): \n%s' %eigenvectors)

# valeurs propres
print('\nValeurs propres de Cov(X): \n%s' %eigenvalues)
```

```python
# étape 3: Arranger les valeurs propres dans l'ordre descroissant

# fait un ensemble de paire (valeur propre, vecteur propre)
val_pairs = [(eigenvalues[index], eigenvectors[:,index]) for index in range(len(eigenvalues))]

# Arrange les paires (valeur propre, vecteur propre) du plus grand au plus petit en fonction de la valeur propre
val_pairs.sort()
val_pairs.reverse()
```

```python
# Variables contenant les valeus propres ordonnées de facon descroissante et 
# les vecteurs propres ordonnée en fonction de leur valeur propre associée 
valeur_sort = [val_pairs[index][0] for index in range(len(eigenvalues))]
vecteurs_sort = [val_pairs[index][1] for index in range(len(eigenvalues))]

types=dataset.columns[:8]
```

```python
# Graphique - Correlation with Outcome

M = dataset.drop(['Outcome'],axis = 1)
M.corrwith(dataset['Outcome']).plot.bar(figsize = (5,5), 
                                        title = "Correlation with Outcome",
                                        fontsize = 10, rot = 90, grid = True)
```

```python
D=np.diag(valeur_sort)
scores = pd.DataFrame(D)
scores.index = ['score'+str(i+1) for i in range(len(eigenvalues))]
scores.head()
```

```python
loadings = pd.DataFrame(vecteurs_sort, columns=cols)
loadings.index = ['PC'+str(i+1) for i in range(len(eigenvalues))]
loadings.head()
```

```python
# Valeurs propres ordonées
print('Valeurs propres en ordre décroissantes: \n%s' %valeur_sort)
```

```python
# Variance cumulée de chaque composante principale
var_comp_sum = np.cumsum(valeur_sort)/sum(valeur_sort)

# Show cumulative proportion of varaince with respect to components
print("Vecteur de proportion des variables explicatives: \n%s" %var_comp_sum)
```

```python
# l'axe des x pour le nombre de composantes principales gardées 
nbr_comp = range(1,len(valeur_sort)+1)
```

```python
# Titre
plt.title('Variances explicatives et composantes gardées')

# nom des axes
plt.xlabel('Composantes principales')
plt.ylabel('Variances explicatives')

# Affichage de l'accumulation des variables explicatives et des composantes principales gardées
plt.scatter(nbr_comp, var_comp_sum,label = "proportion des valeurs explicatives cumulées")
plt.bar(nbr_comp, valeur_sort/sum(valeur_sort),label = "proportion des valeurs explicatives")
plt.legend(loc = 'best')

plt.show()
```

```python
# étape 4: affichage des données en 2d

# garde les deux premières composantes principales
# Réduction de la taille de la taille de la matrice des vecteurs propres ordonées
V_reduit = np.transpose(vecteurs_sort[0:2])

# Les données projetées en 2D seront une matrice n x 2 avec n = nombre d'échantillion/individu
Proj_2D = np.dot(X_std,V_reduit)

# Affiche la projection des données en 2D (teste negatif pour le diabète)
negative = plt.scatter(Proj_2D[:,0][y == 0], Proj_2D[:,1][y == 0])

# Affiche la projection des données en 2D (teste positif pour le diabète)
positive = plt.scatter(Proj_2D[:,0][y == 1], Proj_2D[:,1][y == 1], color = "red")

plt.plot([-6,6],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-6,6],color='silver',linestyle='-',linewidth=1)

# Titre
plt.title('Données réduit en 2D')

# axes
plt.ylabel('Composante principale 2')
plt.xlabel('Composante principale 1')

plt.legend([negative,positive],["Pas de diabète", "Ont le diabète"])

plt.show()
```

# Méthode n°2 pour calculer les PCA.
# Calculer PCA en utilisant SVD

```python
U,singularValues,Vt = np.linalg.svd(X_std,full_matrices= False) #full_matrices=False nous donne pour X une (mxn) : U(mxk), singularValues(kxk), V (kxn) 
D=np.diag(singularValues)
```

```python
#les colonnes de U sont les vecteurs propres de X(X.T)
print('\nU: \n%s' %U)
```

```python
#Les lignes de V sont les vecteurs propres de (X.T)X
print('\nVt: \n%s' %Vt)
```

```python
#Les valeurs propres sont les carrés des valeurs de Sigma
print('singularValues: \n%s' %singularValues)
```

```python
# Variance cumulée de chaque composante principale
var_comp_sum = np.cumsum(singularValues**2)/sum(singularValues**2)

# Show cumulative proportion of varaince with respect to components
print("Vecteur de proportion des variables explicatives: \n%s" %var_comp_sum)
```

```python
# l'axe des x pour le nombre de composantes principales gardées 
nbr_comp = range(1,len(singularValues)+1)

# Titre
plt.title('Variances explicatives et composantes gardées')

# nom des axes
plt.xlabel('Composantes principales')
plt.ylabel('Variances explicatives')

# Affichage de l'accumulation des variables explicatives et des composantes principales gardées
plt.scatter(nbr_comp, var_comp_sum,label = "proportion des valeurs explicatives cumulées")
plt.bar(nbr_comp, (singularValues**2)/sum(singularValues**2),label = "proportion des valeurs explicatives")
plt.legend(loc = 'best')

plt.show()
```

```python
# étape 4: affichage des données en 2d
# garde les deux premières composantes principales

# Les données projetées en 2D seront une matrice n x 2 avec n = nombre d'échantillion/individu

# Matrice composantes principales / individus
Proj_2D= np.dot(U[:,0:2],D[0:2,0:2])

plt.figure(1)
# Affiche la projection des données en 2D (teste negatif pour le diabète)
negative = plt.scatter(Proj_2D[:,0][y == 0], Proj_2D[:,1][y == 0])

# Affiche la projection des données en 2D (teste positif pour le diabète)
positive = plt.scatter(Proj_2D[:,0][y == 1], Proj_2D[:,1][y == 1], color = "red")

# axes
plt.ylabel('Composante principale 2')
plt.xlabel('Composante principale 1')

plt.legend([negative,positive],["Pas de diabète", "Ont le diabète"])


plt.plot([-6,6],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-6,6],color='silver',linestyle='-',linewidth=1)

# Titre
plt.title('Données réduit en 2D')

plt.show()
```
